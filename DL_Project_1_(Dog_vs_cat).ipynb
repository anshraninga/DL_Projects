Resizing all images

```python
#creating a directory for resized images

import os
os.makedirs('/content/dataset/resized', exist_ok=True)
```

```python
original_folder = '/content/dataset/train/'
resized_folder = '/content/dataset/resized/'


for i in range(2000):

  filename = os.listdir(original_folder)[i]
  img_path = os.path.join(original_folder, filename)
  img = Image.open(img_path)
  img = img.resize((224, 224))
  img.save(os.path.join(resized_folder, filename))
  img = img.convert('RGB')
  img.save(os.path.join(resized_folder, filename))

  newImgPath = os.path.join(resized_folder, filename)
  newImg = Image.open(newImgPath)
  newImg.show()
```

Creating labels for resized images of dogs and cats

```python
#Cat--0, Dog--1
#creating a for loop to assign labels

file_names = os.listdir("/content/dataset/resized")

labels = []
for img_file in file_names:
  if img_file[0:3] == 'dog':
    labels.append(1)
  elif img_file[0:3] == 'cat':
    labels.append(0)
```

```python
#converting all the resized images to numpy arrays

import cv2
import glob
```

```python
image_directory = '/content/dataset/resized'
image_extension = ['png', 'jpg']
```

```python
files = []

# Correcting the glob pattern to look inside the directory
[files.extend(glob.glob(os.path.join(image_directory, '*.' + e))) for e in image_extension]

dog_cat_images = np.array([cv2.imread(file) for file in files])
```

Train test split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#scalling the data

X_train_scaled = X_train / 255
X_test_scaled = X_test / 255
```

```python
print(X_train_scaled)
```

Building the neural network

```python
import os
# Force TensorFlow to use the legacy Keras 2
os.environ["TF_USE_LEGACY_KERAS"] = "1"
```

```python
import tensorflow as tf
import tensorflow_hub as hub
import tf_keras as keras
```

```python
mobilenet_model = 'https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4'
```

```python
pretrained_model = hub.KerasLayer(mobilenet_model, input_shape=(224, 224, 3), trainable=False)
```

```python
import tensorflow as tf

# Load MobileNetV2 directly from Keras applications
# include_top=False removes the final classification layer (acting as a feature vector)
# pooling='avg' adds a GlobalAveragePooling layer so the output is a 1D vector
base_model = tf.keras.applications.MobileNetV2(
    input_shape=(224, 224, 3),
    include_top=False,
    weights='imagenet',
    pooling='avg'
)

base_model.trainable = False
num_of_classes = 2

model = tf.keras.Sequential([
    base_model,
    tf.keras.layers.Dense(num_of_classes)
])

model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['acc']
)
```

```python
model.fit(X_train_scaled, y_train, epochs=5)
```

```python
score, acc = model.evaluate(X_test_scaled, y_test)
print('Test score:', score)
print('Test accuracy:', acc)
```

Predictive System

```python
input_image_path = input("Enter the path to the input image: ")

input_image = cv2.imread(input_image_path)

cv2_imshow(input_image)

input_image_resized = cv2.resize(input_image, (224, 224))

input_image_scaled = input_image_resized / 255

input_image_reshaped = np.reshape(input_image_scaled, [1, 224, 224, 3])
input_prediction = model.predict(input_image_reshaped)
input_pred_label = np.argmax(input_prediction)

if input_pred_label == 0:
  print("The image is a cat")
else:
  print("The image is a dog")
```
